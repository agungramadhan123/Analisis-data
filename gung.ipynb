```python
plt.figure(figsize=(10,5))
hitung_korelasi=df.corr(method='pearson')
hitung_korelasi
sns.heatmap(hitung_korelasi, annot=True)
plt.title('korelasi matriks of the dataset')
plt.show()
```

```python
age = df['Age']
pregnancies = df['Pregnancies']

sns.scatterplot(x=age, y=pregnancies, hue=df['Outcome'])
plt.show()
```

Data yang diberikan dari grafik terlihat bahwa sebaran pasien yang terkena diabetes cenderung menyebar dari semua range data umur.
Ini menunjukkan bahwa tidak ada informasi yang kuat atau relasi yang kuat mengenai `Age` dan `Pregnancies`.


```python
sns.scatterplot(x=df['Outcome'], y=df['Glucose'])
plt.show()
```

Tidak didapatkan faktor informasi yang terkait antara `Outcome` dengan level `Glucose`


```python
print(f"Banyak pasien yang tidak terkena diabetes: {len(df[df['Outcome'] == 0])}")
print(f"Banyak pasien yang terkena diabetes: {len(df[df['Outcome'] == 1])}")
```

```python
df['Outcome'].value_counts()
```

```python
proporsi_diabetes = df['Outcome'].value_counts()

plt.pie(proporsi_diabetes.values, autopct="%.1f", labels= proporsi_diabetes.index)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.histplot(data=df, x='Age', hue='Outcome', multiple='stack')
plt.title('Distribusi data Umur berdasarkan Outcome')
plt.show()
```

Bahwa range 20 - 30 merupakan puncak dari pasien yang terkena diabetes. Saran yang bisa dikasih yaitu: Lebih memerhatikan pasien yang berumur di range tersebut.


```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['Insulin'], y=df['SkinThickness'], hue=df['Outcome'])

plt.title('Skin Thickness & Insulin Relation')
plt.show()
```

Semakin Tinggi `SkinThickness` dan semakin tinggi level `Insulin` berarti kemungkinan bahwa pasien tersebut terkena Diabetes berdasarkan hasil grafik diatas.


```python
insuline= df['Insulin']
glokosa=df['Glucose']
sns.scatterplot(x=insuline, y=glokosa, hue=df['Outcome'])
plt.show()
```

data yang di dapatkan bahwa bahwa sebaran di atas bahwa tinggin nya glokosa dan tinggi nya insulin tubuh pasien dapat menyebabkan diabetes
