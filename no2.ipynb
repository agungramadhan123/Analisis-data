```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
#inisiasi variabel
x=0
y=0 
v0=20
sin=45
hitung_sin=(sin/180)*(2 * np.pi)#konversi ke radians
g=-9.8
t=0
dt=0.3
x_arr = [x]
y_arr = [y]
t_arr = [t]
vx = v0 * np.cos(hitung_sin)
vy = v0 * np.sin(hitung_sin)
```

```python
while y >= 0:
 vy += g*dt
 y += vy*dt
 x += vx*dt
 t += dt
 if y < 0:
    break
 # store
 x_arr.append(x)
 y_arr.append(y)
 t_arr.append(t)
```

```python
# numerical solution
# total time
t_tot_num = t_arr[-1]
# range
range_num = x_arr[-1]
# max height
h_max_num = np.max(y_arr)
```

```python
v_ex_arr=[0]
v_ey_arr=[0]
for t in t_arr:
    #hitung gerak peluru solusi numerik
    v_ex = x*(t)+(vx*dt)
    v_ey=y*(t)+(vy*dt)
v_ex_arr.append(v_ex)
v_ey_arr.append(v_ey)


# total time
t_tot_ex = (2 * v0 * np.sin(hitung_sin))/-g
# range
range_ex = v0 * np.cos(hitung_sin) * t_tot_ex
# max height
h_max_ex = (v0**2 * np.sin(hitung_sin)**2) / (-2 * g)
```

```python
plt.figure()
plt.plot(x_arr, y_arr, c='yellow', 
label='numerik')
plt.plot(v_ex_arr, v_ex_arr, c='blue', 
label='analitik')
plt.axhline(c='black')
plt.axvline(c='black')
plt.legend()
plt.title('Solusi numerik vs solusi analitik')
plt.show()
```
