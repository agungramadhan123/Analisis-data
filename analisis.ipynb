```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(10,5))
hitung_korelasi=numeric_df.corr(method='pearson')
hitung_korelasi
sns.heatmap(hitung_korelasi,annot=True)
plt.title('korelasi matriks of dataset')
plt.show()
```
